```python
import pandas as pd

# Load dataset directly from UCI repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Column names (since the file doesn’t have headers)
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

df = pd.read_csv(url, names=columns)

df.head()

```

```python
print("Shape of dataset:", df.shape)          # rows, columns
print("\nColumn Names:", df.columns.tolist()) # column names

print("\nInfo:")
print(df.info())                              # data types

print("\nSummary Statistics:")
print(df.describe())                          # basic stats

print("\nUnique species:", df['species'].unique())
```
